```python
import pandas as pd
import numpy as np
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

import ipywidgets as widgets
from IPython.display import display
```

```python
df = pd.read_csv('/content/sample_data/blr_synthetic_150k_real_locs_updated.csv', on_bad_lines='skip', engine='python')

# IMPORTANT: copy before encoding
df_original = df.copy()

print("Dataset Loaded Successfully")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_location = LabelEncoder()
le_amenities = LabelEncoder()

# fit on ORIGINAL TEXT columns
le_location.fit(df_original['Location'])
le_amenities.fit(df_original['Amenities'])

# transform model dataframe
df['Location'] = le_location.transform(df['Location'])
df['Amenities'] = le_amenities.transform(df['Amenities'])
```

```python
df.drop(
    columns=['PropertyID','TotalFloors','Parking','FurnishingStatus'],
    inplace=True,
    errors='ignore'
)
```

```python
X = df.drop('Price', axis=1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

model = RandomForestRegressor(
    n_estimators=500,
    max_depth=20,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("R2 Score :", r2_score(y_test, y_pred))
print("MAE      :", mean_absolute_error(y_test, y_pred))
```

```python
y_pred = model.predict(X_test)

print("R2 Score :", r2_score(y_test, y_pred))
print("MAE :", mean_absolute_error(y_test, y_pred))
```

```python
max_bhk = int(df_original['BHK'].max())
max_bath = int(df_original['Bath'].max())
max_balcony = int(df_original['Balcony'].max())
max_age = int(df_original['PropertyAgeYears'].max())
max_floor = int(df_original['FloorNumber'].max())

location = widgets.Dropdown(
    options=sorted(df_original['Location'].unique()),
    description='Location'
)

bhk = widgets.IntSlider(min=1,max=max_bhk,value=2,description='BHK')
sqft = widgets.IntText(value=1200,description='Sqft')
bath = widgets.IntSlider(min=1,max=max_bath,value=2,description='Bath')
balcony = widgets.IntSlider(min=0,max=max_balcony,value=1,description='Balcony')
age = widgets.IntSlider(min=0,max=max_age,value=5,description='Age')
floor = widgets.IntSlider(min=0,max=max_floor,value=1,description='Floor')

amenities = widgets.Dropdown(
    options=sorted(df_original['Amenities'].unique()),
    description='Amenities'
)

button = widgets.Button(description='Predict Price',button_style='success')
output = widgets.Output()

def predict_house_price(b):
    with output:
        output.clear_output()

        loc_val = le_location.transform([location.value])[0]
        ame_val = le_amenities.transform([amenities.value])[0]

        new_house = pd.DataFrame([[loc_val,bhk.value,sqft.value,bath.value,
                                   balcony.value,age.value,floor.value,
                                   ame_val]],
        columns=[
            'Location','BHK','TotalSqft','Bath',
            'Balcony','PropertyAgeYears',
            'FloorNumber','Amenities'
        ])

        price = model.predict(new_house)[0]

        print("Predicted House Price =", round(price,2), "Lakhs")


button.on_click(predict_house_price)

display(location,bhk,sqft,bath,balcony,age,floor,amenities,button,output)
```

```python
# Run this in Colab cell

%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor

# -----------------------------------
# PAGE SETTINGS
# -----------------------------------
st.set_page_config(
    page_title="House Price Prediction",
    page_icon="🏠",
    layout="wide"
)

# -----------------------------------
# LOAD DATA
# -----------------------------------
df = pd.read_csv('/content/sample_data/blr_synthetic_150k_real_locs_updated.csv')

df_original = df.copy()

# -----------------------------------
# ENCODING
# -----------------------------------
le_location = LabelEncoder()
le_amenities = LabelEncoder()

le_location.fit(df_original['Location'])
le_amenities.fit(df_original['Amenities'])

df['Location'] = le_location.transform(df['Location'])
df['Amenities'] = le_amenities.transform(df['Amenities'])

# Remove unwanted columns
df.drop(
    columns=['PropertyID','TotalFloors','Parking','FurnishingStatus'],
    inplace=True,
    errors='ignore'
)

# -----------------------------------
# TRAIN MODEL
# -----------------------------------
X = df.drop('Price', axis=1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

model = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

# -----------------------------------
# HEADER
# -----------------------------------
st.markdown(
"""
# 🏠 House Price Prediction
### AI Powered Real Estate Price Predictor
---
"""
)

# -----------------------------------
# SIDEBAR INPUTS
# -----------------------------------
st.sidebar.header("Enter Property Details")

location = st.sidebar.selectbox(
    "Location",
    sorted(df_original['Location'].unique())
)

bhk = st.sidebar.slider(
    "BHK", 1,
    int(df_original['BHK'].max()),
    2
)

sqft = st.sidebar.number_input(
    "Total Sqft",
    min_value=300,
    max_value=10000,
    value=1200
)

bath = st.sidebar.slider(
    "Bath",
    1,
    int(df_original['Bath'].max()),
    2
)

balcony = st.sidebar.slider(
    "Balcony",
    0,
    int(df_original['Balcony'].max()),
    1
)

age = st.sidebar.slider(
    "Property Age",
    0,
    int(df_original['PropertyAgeYears'].max()),
    5
)

floor = st.sidebar.slider(
    "Floor Number",
    0,
    int(df_original['FloorNumber'].max()),
    1
)

amenities = st.sidebar.selectbox(
    "Amenities",
    sorted(df_original['Amenities'].unique())
)

# -----------------------------------
# PREDICTION BUTTON
# -----------------------------------
if st.sidebar.button("Predict Price"):

    loc_val = le_location.transform([location])[0]
    ame_val = le_amenities.transform([amenities])[0]

    new_house = pd.DataFrame(
        [[loc_val,bhk,sqft,bath,balcony,age,floor,ame_val]],
        columns=[
            'Location','BHK','TotalSqft','Bath',
            'Balcony','PropertyAgeYears',
            'FloorNumber','Amenities'
        ]
    )

    price = model.predict(new_house)[0]

    st.success(f"💰 Predicted Price = ₹ {round(price,2)} Lakhs")
    st.info(f"Approx ₹ {round(price/100,2)} Crore")

# -----------------------------------
# MAIN PAGE CONTENT
# -----------------------------------
st.subheader("Why This Project?")
st.write(
"""
This machine learning system predicts house prices using:

- Location
- BHK
- Total Square Feet
- Bathrooms
- Balcony
- Property Age
- Floor Number
- Amenities
"""
)

st.subheader("Model Used")
st.write("Random Forest Regressor")
```
