```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

import ipywidgets as widgets
from IPython.display import display
```

```python
df = pd.read_csv('/content/sample_data/blr_synthetic_150k_real_locs_updated.csv', on_bad_lines='skip', engine='python')

print("Dataset Loaded Successfully")
df.head()
```

```python
df_original = df.copy()

le_location = LabelEncoder()
le_amenities = LabelEncoder()

df['Location'] = le_location.fit_transform(df['Location'])
df['Amenities'] = le_amenities.fit_transform(df['Amenities'])
```

```python
df = df.drop(columns=[
    'PropertyID',
    'TotalFloors',
    'Parking',
    'FurnishingStatus'
], errors='ignore')

df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("R2 Score :", r2_score(y_test, y_pred))
print("MAE      :", mean_absolute_error(y_test, y_pred))
```

```python
max_bhk = int(df_original['BHK'].max())
max_bath = int(df_original['Bath'].max())
max_balcony = int(df_original['Balcony'].max())
max_age = int(df_original['PropertyAgeYears'].max())
max_floor = int(df_original['FloorNumber'].max())
```

```python
location = widgets.Dropdown(
    options=sorted(df_original['Location'].unique()),
    description='Location'
)

bhk = widgets.IntSlider(min=1, max=max_bhk, value=2, description='BHK')

sqft = widgets.IntText(value=1200, description='Sqft')

bath = widgets.IntSlider(min=1, max=max_bath, value=2, description='Bath')

balcony = widgets.IntSlider(min=0, max=max_balcony, value=1, description='Balcony')

age = widgets.IntSlider(min=0, max=max_age, value=5, description='Age')

floor = widgets.IntSlider(min=0, max=max_floor, value=1, description='Floor')

amenities = widgets.Dropdown(
    options=sorted(df_original['Amenities'].unique()),
    description='Amenities'
)

button = widgets.Button(description="Predict Price")
output = widgets.Output()
```

```python
def predict_house_price(b):
    with output:
        output.clear_output()

        loc_val = le_location.transform([location.value])[0]
        ame_val = le_amenities.transform([amenities.value])[0]

        new_house = pd.DataFrame([[loc_val, bhk.value, sqft.value, bath.value,
                                   balcony.value, age.value, floor.value,
                                   ame_val]],
        columns=[
            'Location','BHK','TotalSqft','Bath','Balcony',
            'PropertyAgeYears','FloorNumber','Amenities'
        ])

        price = model.predict(new_house)[0]

        print("Predicted House Price =", round(price,2), "Lakhs")

button.on_click(predict_house_price)

display(
    location,
    bhk,
    sqft,
    bath,
    balcony,
    age,
    floor,
    amenities,
    button,
    output
)
```
